# Handling Missing observation

```python
from sklearn.impute import SimpleImputer
msj = SimpleImputer(missing_values= np.nan, strategy = 'most_frequent')
msj.fit(x)
x = msj.transform(x)
```

```python
x
```

# Splittting data into train and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle=True)
```

```python
#Scaling
```

```python
from sklearn import preprocessing
x_train = preprocessing.normalize(x_train)
x_test = preprocessing.normalize(x_test)
print(x_train)
```

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

```python
x_test
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric = 'euclidean', p = 7)
knn.fit(x_train, y_train)
#default measure minkowski
```

```python
y_pred = knn.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = knn, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

# Artificial Neural Network

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size = 32, epochs = 100)
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
